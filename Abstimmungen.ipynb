```python
def get_results_standard(vorlage):
    results =pd.DataFrame([data[vorlage]['zaehlkreise'][0]['resultat'],
                 data[vorlage]['zaehlkreise'][1]['resultat'],
                 data[vorlage]['zaehlkreise'][2]['resultat'],
                 data[vorlage]['zaehlkreise'][3]['resultat'],
                 data[vorlage]['zaehlkreise'][4]['resultat'],
                 data[vorlage]['zaehlkreise'][5]['resultat'],
                 data[vorlage]['zaehlkreise'][6]['resultat'],
                 data[vorlage]['zaehlkreise'][7]['resultat'],
                 data[vorlage]['zaehlkreise'][8]['resultat']])
    results['ID'] = [re.sub('Zürich Kreis[e]* ', '', kreis['geoLevelname']).strip() for kreis in data[0]['zaehlkreise']]
    results = results[['ID','jaStimmenInProzent','jaStimmenAbsolut','neinStimmenAbsolut','stimmbeteiligungInProzent']]
    results.columns = ['ID', 'C','ja_absolut','nein_absolut','stimmbeteiligung']
    results  = results.round({'C':1,'stimmbeteiligung':1})
    return results
```
